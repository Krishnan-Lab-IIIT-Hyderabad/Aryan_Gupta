# Make 1-D block and spring using Hamiltonian & equation of velocity verlet

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

s = 1000
m = 1
t = 0.01
k = 10
vi = 4

X = np.zeros(s)
V = np.zeros(s)
H = np.zeros(s)

X[0] = 0
V[0] = vi

x1 = np.linspace(0, 1, s)

for i in range(s-1):
    cur_x = X[i]
    cur_v = V[i]
    a1 = -(k*cur_x)/m
    cur_xx = cur_x + t*cur_v + (t*t*a1)/2
    X[i+1] = cur_xx
    a2 = -(k*cur_xx)/m
    cur_vv = cur_v + ((a1+a2)*t)/2
    V[i+1] = cur_vv
    
H = (m*V*V)/2 + (k*X*X)/2
```

```python
plt.plot(x1, V)
plt.plot(x1, X)
plt.plot(x1, H)
```

```python
ax = plt.axes(projection ='3d')
ax.plot3D(X, V, H, 'green')
```
